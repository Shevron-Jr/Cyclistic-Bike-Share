#### BUSINESS TASK:
What is the problem?

1.How do annual members and casual riders use Cyclistic bikes differently?

#### DATASET:
The data is on an AWS server where it is easily downloadable and named correctly. I downloaded the previous 11 months data for the year 2022, and stored it locally for the next steps in the analysis processes. The data is current, cited, original but not reliable. There are some null values and outliers in the data.

Note that data-privacy issues prohibits me from using riders’ personally identifiable information. This means that you won’t be able to connect pass purchases to credit card numbers to determine if casual riders live in the Cyclistic service area or if they have purchased multiple single passes.

The files are saved in comma-seperated values (CSV).

Python is used to analyze this data

Link to previous 12 months of Cyclistic Trip Data: https://divvy-tripdata.s3.amazonaws.com/index.html  

```python
# importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
import datetime 
import matplotlib.pyplot as plt
```

```python
# Importing the downloaded datasets
data_1 = pd.read_csv('01-january-2022.csv')
data_2 = pd.read_csv('02-february-2022.csv')
data_3 = pd.read_csv('03-march-2022.csv')
data_4 = pd.read_csv('04-april-2022.csv')
data_5 = pd.read_csv('05-may-2022.csv')
data_6 = pd.read_csv('06-june-2022.csv')
data_7 = pd.read_csv('07-july-2022.csv')
data_8 = pd.read_csv('08-august-2022.csv')
data_9 = pd.read_csv('09-september-2022.csv')
data_10 = pd.read_csv('10-october-2022.csv')
data_11 = pd.read_csv('11-november-2022.csv')
```

```python
data_5.head()
```

```python
# checking the info of each monthly data, and checking for any null values
#It was discovered there was no null values in all 11 months data for 2022
data_1.info()
```

```python
# Merging the data
bike_data = pd.concat([data_1, data_2, data_3, data_4, data_5, data_6, data_7, data_8, data_9, data_10, data_11])
bike_data
```

```python
bike_data.head()
```

```python
bike_data.info()
```

The merged data set for the 11 months (January to November, 2022) conists of 5,485,911 rows and 13 columns in total.

```python
# Creating a backup of the original merged data
bike_data.to_csv('bike_data1')
```

```python
# changing ride_type and member columns to category data types
bike_data['rideable_type'] = bike_data.rideable_type.astype('category')
bike_data['member_casual'] = bike_data.member_casual.astype('category')
```

```python
bike_data.info()
```

```python
# dropping the lat and lng columns from the datasets
bike_data.drop(['start_lat', 'start_lng', 'end_lat', 'end_lng'], axis = 1, inplace = True)
```

```python
bike_data.info
bike_data.head()
```

```python
# changing start time and end time to a datetime datatype
bike_data['started_at'] = pd.to_datetime(bike_data.started_at)
bike_data['ended_at'] = pd.to_datetime(bike_data.ended_at)
```

```python
bike_data
bike_data.info()
```

```python
# finding the missing values
# start station nae, the id, end station name and the id all has over 800,000 missing values
bike_data.isnull().sum()
```

```python
# adding a new column of the ride_length, (ended_at - started_at) to calculate the time travelled
bike_data['ride_length'] = bike_data['ended_at'] - bike_data['started_at']
bike_data['ride_length']
```

```python
bike_data.info()
```

```python
# splitting and extracting the date from the 'started_at' column
bike_data['start_date'] = bike_data.started_at.dt.date
bike_data.start_date.head()
```

```python
# converting 'day_of_week' into a datetime dtype
bike_data['start_date'] = pd.to_datetime(bike_data.start_date)
bike_data.dtypes
```

```python
bike_data.info()
```

```python
# finding the weekday of the date
bike_data['day_name'] = bike_data['start_date'].dt.day_name()
bike_data['day_name'].unique()
```

```python
# changing 'day_name' to category
bike_data['day_name'] = bike_data.day_name.astype('category')
bike_data.info()
```

```python
bike_data['year'] = pd.DatetimeIndex(bike_data['start_date']).year
bike_data['month'] = pd.DatetimeIndex(bike_data['start_date']).month
```

```python
bike_data.head()
```

```python
# sort the dataframe by ascending using 'started_at'
bike_data.sort_values(by = ['started_at'], inplace = True, ascending = True)
```

```python
bike_data.head()
```

```python
bike_data.isnull().sum()
```

```python
# drop all rows with NAN values 
bike_data.dropna()
bike_data.dropna(inplace = True)
```

```python
# to save the cleaned file
bike_data.to_csv('bike_data_cleaned.csv')
```
